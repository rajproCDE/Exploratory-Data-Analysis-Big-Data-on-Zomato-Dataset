# **1. What are the Available Cuisines Across all Restaurants Listed in Zomato?**

```python
from pprint import pprint

cuisines_total = df['Cuisines']
texts = []
hash_map = set()
cuisines_total.dropna(inplace=True)
for i in cuisines_total:
    for j in str(i).split(', '):
        if j not in hash_map:
            texts.append(j.lower())
            hash_map.add(j)
pprint(texts)
```

# **2. Let's see how ratings are distributed?**

> If people are passionately disliking the restaurants on the app there is a good chance they will leave a negative rating. therefore let's see how well the restaurants are faring based on ratings.



> First we need to filter out ratings outside of India and separate the columns representing ratings. The country code of India is given to be 1.





```python
ratingdf = df.loc[df['Country Code'] == 1].groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
ratingdf
```

```python
ratingdf.rename(columns={0:'Count'}, inplace=True)
ratingdf
```

```python
col = ['white', 'red', 'orange', 'yellow', 'green', 'darkgreen']
sns.set_style('darkgrid')
sns.barplot(x='Aggregate rating', y='Count', hue='Rating text', data=ratingdf, palette = col)
plt.rcParams['figure.figsize'] = (10, 10)
```
